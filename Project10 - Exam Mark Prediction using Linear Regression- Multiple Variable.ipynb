#### Project Title: Exam Mark Prediction using Linear Regression - Multiple Variables

#### Objectives: Predicting Exam mark based on hours, age and Internet availability

#### *Importing Libraries*

```python
import pandas as pd
from sklearn.linear_model import LinearRegression 
```

#### *Loading Dataset*

```python
dataset = pd.read_csv('data1.csv')
dataset
```

#### *Summarizing Dataset*

```python
print(dataset.shape)
print(dataset.head(5))
```

#### *Finding & Removing NA values from our Features X*

```python
dataset.columns[dataset.isna().any()]
```

```python
dataset.hours = dataset.hours.fillna(dataset.hours.mean())
```

#### *Recheck for NA values*

```python
dataset.columns[dataset.isna().any()]
```

#### *Segregate Dataset into Input X & Output Y*

```python
#X = dataset.iloc[:, :-1].values

X=dataset.drop('marks', axis='columns')
print(X.shape)
print(X)
```

```python
#Y = dataset.iloc[:, -1].values

Y=dataset.marks
Y
```

#### *Training Dataset using Linear Regression*

```python
model = LinearRegression()
model.fit(X,Y)
```

#### *Predicted mark using custom values*

```python
a=[[6.2,20,1]]
PredictedmodelResult = model.predict(a)
print(PredictedmodelResult)
```

```python
import matplotlib.pyplot as plt
#plt.scatter(X,Y, color="red")
plt.plot(X.age,Y)
plt.title("Linear Regression")
plt.xlabel("Age")
plt.ylabel("marks")
plt.grid()
plt.show()
```
